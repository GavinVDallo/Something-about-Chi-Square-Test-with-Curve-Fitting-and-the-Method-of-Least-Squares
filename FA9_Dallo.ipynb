```python
observed = [24, 108, 95, 23]
expected = [30, 90, 90, 30]

# Calculate Chi-Square: sum of ((O - E)^2 / E)
chi_sq = sum((o - e)**2 / e for o, e in zip(observed, expected))

print("The coin is tossed 240 times and observing each time a coin shows up heads with a significance level of 0.05")

print(f"Calculated Chi-Square Statistic: {chi_sq:.4f}")

# Critical value for df=3 at alpha=0.05 is 7.815
critical_value = 7.815

if chi_sq > critical_value:
    print("Reject Null Hypothesis: Coins are likely not fair.")
else:
    print("Fail to Reject: No significant evidence that coins are unfair.")
```

```python
import math

print("looking for the data result about the relationship between hair and eye color selected from 200 students")

# Data from Table 3
a, b = 49, 25  # Blue eye (Blonde, Not blonde)
c, d = 30, 96  # Not blue eye (Blonde, Not blonde)
N = a + b + c + d

# Row and Column Totals
r1, r2 = a + b, c + d
c1, c2 = a + c, b + d

# 1. Calculation WITHOUT Yates' Correction
# Shortcut formula for 2x2: N * (ad - bc)^2 / (r1 * r2 * c1 * c2)
numerator = N * (a * d - b * c)**2
denominator = r1 * r2 * c1 * c2
chi_sq = numerator / denominator

# Coefficient of Contingency (C)
c_coeff = math.sqrt(chi_sq / (chi_sq + N))

# 2. Calculation WITH Yates' Correction
# Shortcut: N * (|ad - bc| - N/2)^2 / (r1 * r2 * c1 * c2)
num_yates = N * (abs(a * d - b * c) - N/2)**2
chi_sq_yates = num_yates / denominator
c_yates = math.sqrt(chi_sq_yates / (chi_sq_yates + N))

# 3. Maximum Coefficient (C_max) for a 2x2 table
c_max = math.sqrt((2 - 1) / 2)

print(f"--- Results ---")
print(f"C (No Yates):     {c_coeff:.4f}")
print(f"C (With Yates):   {c_yates:.4f}")
print(f"C Max:           {c_max:.4f}")
```

```python
import math

# Sample data from mtcars (Weight in 1000 lbs, MPG)
wt = [2.62, 2.875, 2.32, 3.215, 3.44, 3.46, 3.57, 3.19, 3.15, 3.44]
mpg = [21.0, 21.0, 22.8, 21.4, 18.7, 18.1, 14.3, 24.4, 22.8, 19.2]
n = len(wt)

print("Looking for the result and analysis about gallons and weight of 32 automobiles")

def get_regression(x_data, y_data):
    sum_x = sum(x_data)
    sum_y = sum(y_data)
    sum_xy = sum(x * y for x, y in zip(x_data, y_data))
    sum_x_sq = sum(x**2 for x in x_data)
    
    # Calculate Slope (m) and Intercept (b)
    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_sq - sum_x**2)
    b = (sum_y - m * sum_x) / n
    return m, b

# (a) mpg as dependent (y), wt as independent (x)
slope_a, intercept_a = get_regression(wt, mpg)

# (b) wt as dependent (y), mpg as independent (x)
slope_b, intercept_b = get_regression(mpg, wt)

print(f"Model (a) [mpg ~ wt]: mpg = {intercept_a:.2f} + ({slope_a:.2f} * wt)")
print(f"Model (b) [wt ~ mpg]: wt = {intercept_b:.2f} + ({slope_b:.2f} * mpg)")

# (d) Interpretation
print(f"\nInterpretation (a): For every 1000 lbs increase, MPG changes by {slope_a:.2f}.")
print(f"Interpretation (b): For every 1 MPG increase, Weight changes by {slope_b:.2f} (1000 lbs).")

```

```python
import numpy as np

# Load the cars dataset (Speed in mph, Distance in ft)
# These are the first several points from the R 'cars' dataset
speed = np.array([4, 4, 7, 7, 8, 9, 10, 10, 10, 11, 12, 12, 13, 14, 15, 16, 17, 18, 19, 20, 22, 24, 25])
dist = np.array([2, 10, 4, 22, 16, 10, 18, 26, 34, 17, 28, 14, 34, 36, 48, 40, 50, 56, 68, 48, 66, 92, 85])

# (b) Fit Quadratic Model: D = a0 + a1*V + a2*V^2
# np.polyfit(x, y, degree)
quad_coeffs = np.polyfit(speed, dist, 2)
a2, a1, a0 = quad_coeffs  # Note: returns highest power first

# (e) Fit Linear Model: D = b0 + b1*V
lin_coeffs = np.polyfit(speed, dist, 1)
b1, b0 = lin_coeffs

print("Finding distance and speed of 50 automobiles")

print(f"Quadratic Model: D = {a0:.2f} + {a1:.2f}V + {a2:.2f}V^2")
print(f"Linear Model:    D = {b0:.2f} + {b1:.2f}V")

# (d) Estimate stopping distance for V = 45 and 80
def estimate(v):
    return a0 + a1*v + a2*(v**2)

print(f"\n--- Predictions (Quadratic) ---")
print(f"At 45 mph: {estimate(45):.2f} feet")
print(f"At 80 mph: {estimate(80):.2f} feet")
```
